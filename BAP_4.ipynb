```python
import pandas as pd
df = pd.read_csv(r'C:\Users\user\Desktop\numpy\multiTimeline.csv')
df
```

```python
df['Week'] = pd.to_datetime(df['Week'])
df
```

```python
df.set_index('Week', inplace=True)
```

```python
# Resample to different time frequencies
# Up sampling with forward fill (FFill)
df_upsampled = df.resample('D').ffill()
```

```python
# Print the results
print("Upsampled with Forward Fill:")
print(df_upsampled.head())
```

```python
# Down sampling with OHLC sampling
df_downsampled = df.resample('M').ohlc()
```

```python

print("\nDownsampled with OHLC Sampling:")
print(df_downsampled.head())
```

```python
# Shift time series forward and backward
# Naive shift
df_shifted_forward = df.shift(1)
df_shifted_backward = df.shift(-1)
```

```python
print("\nShifted Time Series (Naive - Forward):")
print(df_shifted_forward.head())
```

```python

print("\nShifted Time Series (Naive - Backward):")
print(df_shifted_backward.head())
```

```python
# Shift using frequency
df_shifted_freq_forward = df.shift(freq='D')
print("\nShifted Time Series (Frequency - Forward):")
print(df_shifted_freq_forward.head())
```

```python
df_shifted_freq_backward = df.shift(freq='-D')
print("\nShifted Time Series (Frequency - Backward):")
print(df_shifted_freq_backward.head())
```

```python
# Compute moving averages or rolling sums
rolling_mean_7 = df.rolling(window=7).mean()
print("\nRolling Mean (7-day Window):")
print(rolling_mean_7.head())
```

```python
rolling_sum_7 = df.rolling(window=7).sum()
print("\nRolling Sum (7-day Window):")
print(rolling_sum_7.head())
```

```python
rolling_mean_30 = df.rolling(window=30).mean()
print("\nRolling Mean (30-day Window):")
print(rolling_mean_30.head())
```
