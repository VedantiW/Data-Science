### Pivot

```python
duplicates = df[df.duplicated(subset='brand', keep=False)]

df = df.drop_duplicates(subset='brand', keep='first')

pivot_df = df.pivot(index='brand', columns='model', values=['colour', 'storage'])
print(pivot_df)
```

1) Groupby

```python
grouped_df = df.groupby(['brand', 'model']).sum().reset_index()
print("\nDataFrame after grouping and aggregation:")
print(grouped_df)
```
